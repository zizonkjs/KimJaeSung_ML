#### [3D 시각화 예시]

```python
# 모듈로딩
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 생성
x = np.linspace(-10, 10, 11)
print(f'[x]{x.shape}\n{x}\n')

X = np.tile(x, (11, 1))
print(f'[x]{X.shape}\n{x}\n')
# tile 함수는 주어진 배열을 반복하여 새로운 배열을 생성합니다.
Y = np.transpose(X)
print(f'[x]{Y.shape}\n{x}\n')
#transpose 함수는 배열의 축(axes)을 교환하여 배열을 변환합니다. 
#일반적으로 행렬의 전치(transpose)를 구하는 데 사용됩니다.

Z = np.random.rand(11, 11)
```

```python
# 시각화 진행
fig = plt.figure(figsize=(10,6))

# 3차원 창 설정 => projection='3d'
ax = fig.add_subplot(111, projection='3d')

# 평면 생성 : 2차원 어레이를 3차원 Surface로 표현
ax.plot_surface(X, Y, Z, color='yellow')
#  3D 표면 플롯을 그리는 함수입니다
ax.set_zlim(-5, 5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```

- [예시2] 

```python
# 모듈로딩
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
## 임의의 데이터 
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
z = [-1, -1, -1, -1]

## 평면 좌표 데이터 생성
# - 좌표 범위 생성
X_data = np.arange(-0.5, 1.5, 0.1)
Y_data = np.arange(-0.5, 1.5, 0.1)

# - 행렬 ==> 좌표값 변환
X, Y = np.meshgrid(X_data, Y_data)
# 여러 1차원 배열로부터 좌표 행렬을 생성하는 데 사용됩니다. 
# 이 함수는 주로 2차원 또는 3차원 공간에서 함수의 좌표 그리드를 생성할 때 유용하며, 
# 특히 plot_surface, contour, imshow 등의 그래프를 그릴 때 자주 사용됩니다.
Z = (float(-5/7) * Y) + (float(-4/7) * X)
```

```python
## 시각화 
fig = plt.figure()
# projection : 객체가 그래프에 투영될 방법 설정
#              None, 'mollweide', 'polar', 'rectilinear', str
ax = fig.add_subplot(111, projection='3d')


# 데이터 출력
ax.plot(x, y, z, 
        linestyle="none", 
        marker="o", mfc="none", markeredgecolor="red")

# 데이터 평면 출력
# stride, cstride : row와 column에서 데이터에서 어느정도 graph 그리는데 사용할지 설정
# alpha : 그래프의 투명도  
# cmap : 색상 설정 color map


ax.plot_surface(X, Y, Z, 
                rstride=4, cstride=4, 
                alpha=0.4, cmap=cm.jet)

plt.show()
```
